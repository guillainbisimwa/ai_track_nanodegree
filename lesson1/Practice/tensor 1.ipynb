```python
import torch
```

```python
def activation(x):
    #signoid function of "x" tensor
    return 1/(1+torch.exp(-x))
```

```python
torch.manual_seed(7)

features = torch.randn((1,5))

weight = torch.rand_like(features)
```

```python
print(features.shape)
print(features)
```

```python
print(weight.shape)
print(weight)
```

```python
bias = torch.randn((1,1))
print(bias)
```

```python
#calculate the output of this nn
y = activation(torch.mm(features, weight.view(5,1)) + bias)
print(y)
```
